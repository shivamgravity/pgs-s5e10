# Preprocessing Script

Process the data to make it digestable for ML models later.

```python
# import required libraries
import pandas as pd
```

```python
# read the training dataset
train = pd.read_csv("../../dataset/train.csv")
```

```python
u_road_type = list(train['road_type'].unique())
u_lighting = list(train['lighting'].unique())
u_weather = list(train['weather'].unique())
u_road_sign_present = list(train['road_signs_present'].unique())
u_public_road = list(train['public_road'].unique())
u_time_of_day = list(train['time_of_day'].unique())
u_holiday = list(train['holiday'].unique())
u_school_season = list(train['school_season'].unique())
```

```python
road_type = []
for x in train['road_type']:
    if x == "urban":
        road_type.append(3)
    elif x == "highway":
        road_type.append(2)
    elif x == "rural":
        road_type.append(1)
    else:
        print("Road Type Error!")
        break
train['road_type'] = road_type
```

```python
lighting = []
for x in train['lighting']:
    if x == "daylight":
        lighting.append(1)
    elif x == "dim":
        lighting.append(2)
    elif x == "night":
        lighting.append(3)
    else:
        print("Lightning Type Error!")
        break
train['lighting'] = lighting
```

```python
weather = []
for x in train['weather']:
    if x == "clear":
        weather.append(1)
    elif x == "foggy":
        weather.append(2)
    elif x == "rainy":
        weather.append(3)
    else:
        print("Weather Type Error!")
        break
train['weather'] = weather
```

```python
road_signs_present = []
for x in train['road_signs_present']:
    if x is True:
        road_signs_present.append(-1)
    elif x is False:
        road_signs_present.append(1)
    else:
        print("Road Sign Type Error!")
        break
train['road_signs_present'] = road_signs_present
```

```python
public_road = []
for x in train['public_road']:
    if x is True:
        public_road.append(1)
    elif x is False:
        public_road.append(-1)
    else:
        print("Public Road Type Error!")
        break
train['public_road'] = public_road
```

```python
time_of_day = []
for x in train['time_of_day']:
    if x == "afternoon":
        time_of_day.append(1)
    elif x == "morning":
        time_of_day.append(2)
    elif x == "evening":
        time_of_day.append(3)
    else:
        print("Time of day Error!")
        break
train['time_of_day'] = time_of_day
```

```python
holiday = []
for x in train['holiday']:
    if x is True:
        holiday.append(1)
    elif x is False:
        holiday.append(-1)
    else:
        print("Holiday Error!")
        break
train['holiday'] = holiday
```

```python
school_season = []
for x in train['school_season']:
    if x is True:
        school_season.append(-1)
    elif x is False:
        school_season.append(1)
    else:
        print("School Error!")
        break
train['school_season'] = school_season
```
